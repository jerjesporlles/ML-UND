```python
#Librerias a utilizar para Machine Learning
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# La data para entrenar y predecir el modelo
dataset = pd.read_excel("dataset.xlsx", sheet_name="data")
```

```python
dataset
```

```python
dataset['Formacion'].unique()
```

```python
# Identidicamos que categorias tenemos en cada columna
print('Categorias de la columna Formacion: ',dataset['Formacion'].unique())
print('Categorias de la columna Tipo de grano: ',dataset['Tipo de grano'].unique())
print('Categorias de la columna Tipo de permeabilidad: ',dataset['Tipo de permeabilidad'].unique())
```

```python
dataset['Formacion'].replace(('A1','A2','A3'),(1,2,3),inplace = True)
```

```python
dataset
```

```python
# Identidicamos que categorias tenemos en cada columna
print('Categorias de la columna Formacion: ',dataset['Formacion'].unique())
print('Categorias de la columna Tipo de grano: ',dataset['Tipo de grano'].unique())
print('Categorias de la columna Tipo de permeabilidad: ',dataset['Tipo de permeabilidad'].unique())
```

```python
dataset['Tipo de grano'].replace(('Grano fino','Grano medio','Grano grueso'),(1,2,3),inplace=True)
dataset['Tipo de permeabilidad'].replace(('alta','Media','Baja'),(1,2,3),inplace=True)
```

```python
dataset
```

```python
variables = dataset.iloc[:,[0,1,2,3,4,5,6]].values
```

```python
variables
```

```python
resultado = dataset.iloc[:,[8]].values
```

```python
resultado
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(variables,resultado,
                                                test_size=0.25,random_state=0)
```

```python
len(x_test)
```

```python
# Estandarizar las escalas
from sklearn.preprocessing import StandardScaler
escala_estandarizada = StandardScaler()
```

```python
x_train= escala_estandarizada.fit_transform(x_train)
x_test = escala_estandarizada.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clasificador = KNeighborsClassifier(n_neighbors=5,metric='minkowski', p = 2)
```

```python
clasificador.fit(x_train,y_train)
```

```python
y_pred = clasificador.predict(x_test)
```

```python
y_pred
```

```python
data_pred = pd.DataFrame(y_pred)
```

```python
data_pred.rename(columns={'y_test':'y_pred'},inplace = True)
data_pred
```

```python
y_test
```

```python
data_test = pd.DataFrame(y_test)
data_test.rename(columns={0:'y_test'},inplace = True)
data_test
```

```python
df_comparacion = pd.concat([data_test, data_pred], axis=1)
df_comparacion
df_comparacion['Diferencia']=df_comparacion.y_test - df_comparacion.y_pred
df_comparacion[30:60]
```

```python
from sklearn.metrics import confusion_matrix
matris_confusion = confusion_matrix(y_test, y_pred)
matris_confusion
```

```python
sc = clasificador.score(x_test,y_test)
sc
```
